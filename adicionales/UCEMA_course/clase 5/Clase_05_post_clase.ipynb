## MÃ©todo mas rÃºstico, iterando

```python
%%time

data = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1', index_col='timestamp')
data.sort_index(inplace=True)
p = [0.2, 0.15, 0.15, 0.5]

pxs = []
for idx, row in data.iterrows():
    px = row.open * p[0] + row.high * p[1] + row.low *p[2] + row.close * p[3]
    pxs.append(px)
    
data['px_OHLC'] = pxs
data
```

```python
data = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1', index_col='timestamp')
data.sort_index(inplace=True)
p = [0.2, 0.15, 0.15, 0.5]

pxs = []
for i in range(len(data)):
    row = data.iloc[i]
    px = row.open * p[0] + row.high * p[1] + row.low *p[2] + row.close * p[3]
    pxs.append(px)
    
data['px_OHLC'] = pxs
data
```
